```python
df = pd.read_csv("data/Data_Entry_2017_v2020.csv")
```

```python
Base.metadata.create_all(bind=engine)
```

```python
picture_schema = {'picture_id': None,
 'patient_id': None,
 'Abnormal': False,
 'Atelectasis': False,
 'Cardiomegaly': False,
 'Effusion': False,
 'Infiltration': False,
 'Mass': False,
 'Nodule': False,
 'Pneumonia': False,
 'Pneumothorax': False,
 'Consolidation': False,
 'Edema': False,
 'Emphysema': False,
 'Fibrosis': False,
 'PleuralThickening': False,
 'Hernia': False,
 'Other': False}
```

```python
try:
    db = SessionLocal()
    i = 0
    for idx, row in df.iterrows():
        pic = picture_schema.copy()
        patient_id = row['Patient ID']
        pic['patient_id'] = patient_id
        pic['picture_id'] = row['Image Index']
        age = row['Patient Age']
        gender = row['Patient Gender']
        findings = row['Finding Labels'].split('|')
        for finding in findings:
            if finding == "No Finding":
                continue
            elif finding == "Pleural Thickening":
                pic['PluralThickening'] = True
            else:
                row[finding] = True
        if i % 1000 == 0:
            print(f'Added {i} images')
        i+= 1
        crud.populate_db(db,pic, patient_id, age, gender)
finally:
    db.close()
```
